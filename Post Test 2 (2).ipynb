```python
import pandas as pd
import numpy as np
```

```python
data_jualan = pd.read_csv("TransactionModel2.csv", encoding="latin1", sep=";")
data_jualan
```

```python
data_jualan.isna().sum()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data_jualan["Item_Name"].fillna(data_jualan["Item_Name"].mode()[0], inplace=True)
```

```python
data_jualan["Item_Qty"].fillna(data_jualan["Item_Qty"].mean(), inplace=True)
```

```python
data_jualan["Item_Price"].fillna(data_jualan["Item_Price"].mean(), inplace=True)
```

```python
data_jualan.loc[data_jualan["Item_Total"].isna(), "Item_Total"] = data_jualan["Item_Qty"]*data_jualan["Item_Price"]
data_jualan
```

```python
data_jualan.info()
```

```python
data_jualan["Supplier_Name"] = data_jualan["Supplier_Name"].astype(str)
```

```python
data_jualan["Supplier_Type"] = data_jualan["Supplier_Type"].astype(str)
```

```python
data_jualan["Warehouse_Name"] = data_jualan["Warehouse_Name"].astype(str)
```

```python
data_jualan["Item_Name"] = data_jualan["Item_Name"].astype(str)
```

```python
data_jualan["Item_Type"] = data_jualan["Item_Type"].astype(str)
```

2. Analisis Penjualan Berdasarkan Supplier

```python
filter_multiple_suppliers = filter_data1[filter_data1["Supplier_Name"].isin(["Kelompok Tani Merdeka", "Kelompok Tani Merah"])]
sales_by_supplier_multiple = filter_multiple_suppliers.groupby("Supplier_Name")["Item_Total"].sum().sort_values(ascending=False)
sales_by_supplier_multiple
```

```python
sales_by_supplier_multiple = filter_data1.groupby("Supplier_Name")["Item_Total"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.bar(sales_by_supplier_multiple.index, 
        sales_by_supplier_multiple.values, 
        color="lightcoral")

plt.xlabel("Supplier Name")
plt.ylabel("Total Sales (Item_Total)")
plt.title("Total Sales per Supplier (All Suppliers)")
plt.xticks(rotation=45,ha="right")

plt.tight_layout()  
plt.subplots_adjust(bottom=0.2)
plt.show()

```

3. Menganalisa Penjualan Berdasarkan Warehouse

```python
sales_by_warehouse = filter_data1.groupby("Warehouse_Name")["Item_Total"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
plt.bar(sales_by_warehouse.index, 
        sales_by_warehouse.values, 
        color="brown")

plt.xlabel("Warehouse Name")
plt.ylabel("Total Sales (Item_Total)")
plt.title("Total Sales per Warehouse")
plt.xticks(rotation=45, ha='right')  

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)

plt.show()
```

4. Analisis Penjualan Anggur

```python
filter_item_grape = filter_data1[filter_data1["Item_Name"] == "Anggur/Grape\xa0"]
sales_by_grape_month = filter_item_grape.groupby("Bulan")["Item_Total"].sum()
```

```python
plt.figure(figsize=(8,6))
plt.bar(sales_by_grape_month.index, 
        sales_by_grape_month.values, 
        color="purple")

plt.xlabel("Month")
plt.ylabel("Total Sales (Item_Total)")
plt.title("Sales of Anggur/Grape per Month")
plt.xticks(rotation=45, ha='right') 

plt.tight_layout()
plt.show()
```

5. Analisis Penjualan Berdasarkan Harga dan Quantity

```python
sales_by_price_qty = filter_data1.groupby(["Item_Price", "Item_Qty"])["Item_Total"].sum().reset_index()
sales_by_price_qty['Price_Quantity'] = sales_by_price_qty['Item_Price'].astype(str) + " / " + sales_by_price_qty['Item_Qty'].astype(str)
```

```python
plt.figure(figsize=(10,6))
plt.bar(sales_by_price_qty["Price_Quantity"], 
        sales_by_price_qty["Item_Total"], 
        color="green")


plt.xlabel("Price / Quantity")
plt.ylabel("Total Sales (Item_Total)")
plt.title("Sales Based on Price and Quantity")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
