```python
#4
df['Tanggal'] = pd.to_datetime(df['Tanggal'], errors='coerce')

df[
    (df['Tanggal'].dt.month == 10) & 
    (df['Tanggal'].dt.year == 2023) & 
    (df["Jenis Produk"] == "HVS") &
    (df['Total'] == df.loc[
        (df['Tanggal'].dt.month == 10) & 
        (df['Tanggal'].dt.year == 2023), 'Total'].min())
]
```
