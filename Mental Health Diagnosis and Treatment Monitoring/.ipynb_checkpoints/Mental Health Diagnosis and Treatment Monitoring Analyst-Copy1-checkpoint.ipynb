Latar Belakang

Kesehatan mental merupakan aspek penting dalam kesejahteraan individu dan masyarakat. Dalam beberapa tahun terakhir, kesadaran akan pentingnya kesehatan mental semakin meningkat, seiring dengan bertambahnya jumlah individu yang mencari diagnosis dan perawatan untuk berbagai gangguan mental, seperti depresi, kecemasan, dan gangguan stres pascatrauma.

Namun, masih terdapat berbagai tantangan dalam mendiagnosis serta memberikan perawatan yang tepat bagi individu dengan masalah kesehatan mental. Faktor-faktor seperti stigma sosial, akses terhadap layanan kesehatan, kondisi ekonomi, serta dukungan keluarga dapat mempengaruhi keputusan seseorang dalam mencari bantuan profesional. Oleh karena itu, analisis terhadap faktor-faktor ini sangat penting untuk memahami pola diagnosis dan perawatan kesehatan mental.

Dataset yang digunakan dalam proyek ini adalah mental_health_diagnosis_treatment_.csv, yang berisi informasi mengenai individu yang telah didiagnosis dengan gangguan kesehatan mental serta faktor-faktor yang mempengaruhi keputusan mereka untuk mencari perawatan. Dataset ini mencakup variabel seperti usia, jenis kelamin, riwayat kesehatan mental, akses terhadap layanan kesehatan, serta pengalaman dalam menerima diagnosis dan perawatan. Data ini dapat digunakan untuk melakukan berbagai analisis, antara lain:

1. Analisis Pola Gejala dan Diagnosis

    Mengidentifikasi hubungan antara tingkat keparahan gejala, kualitas tidur, tingkat stres, dan skor suasana hati dengan diagnosis kesehatan mental pasien.

2. Analisis Efektivitas Perawatan

    Menganalisis pengaruh durasi perawatan, jenis terapi, pengobatan, tingkat kepatuhan terhadap perawatan, dan aktivitas fisik terhadap hasil perawatan pasien.

3. Analisis Kondisi Emosional dan Tren Perubahan

    Menganalisis hubungan antara kondisi emosional yang terdeteksi oleh AI, skor suasana hati, dan tingkat stres, serta melihat pola perubahan kondisi emosional pasien selama perawatan.

Selain analisis eksploratif, dataset ini juga dapat digunakan untuk membangun model prediktif yang dapat membantu dalam:

1. Prediksi Diagnosis Kesehatan Mental

   Mengklasifikasikan pasien berdasarkan kondisi kesehatan mental mereka berdasarkan faktor-faktor dalam dataset.

2. Prediksi Hasil Perawatan

    Menganalisis kemungkinan keberhasilan atau kegagalan perawatan berdasarkan faktor seperti durasi terapi, jenis pengobatan, dan tingkat kepatuhan pasien terhadap perawatan.



Penjelasan Per Kolom 

1. Patient ID = Identifikasi unik untuk setiap pasien.
2. Age = Usia pasien dalam tahun.
3. Gender = Jenis kelamin pasien (Male/Female).
4. Diagnosis = Kondisi kesehatan mental yang dialami pasien (misalnya, Anxiety, Depression).
5. Symptom Severity (1-10) = Tingkat keparahan gejala yang dialami pasien, dalam skala 1-10.
6. Mood Score (1-10) = Skor suasana hati pasien selama perawatan, dalam skala 1-10.
7. Sleep Quality (1-10) = Kualitas tidur pasien yang dilaporkan sendiri, dalam skala 1-10.
8. Physical Activity = Jumlah jam aktivitas fisik pasien per minggu.
9. Medication = Jenis obat yang diresepkan untuk pasien (misalnya, SSRIs, Antidepressants).
10. Therapy Type = Jenis terapi yang diterima pasien (misalnya, CBT, DBT).
11. Treatment Start Date = Tanggal dimulainya perawatan pasien.
12. Treatment Duration = Lama perawatan dalam minggu.
13. Stress Level (1-10) = Tingkat stres pasien yang dilaporkan sendiri, dalam skala 1-10.
14. Outcome = Hasil dari perawatan yang dijalani pasien (misalnya, Improved, Deteriorated).
15. Treatment Progress (1-10) = Kemajuan yang dibuat pasien selama perawatan, dalam skala 1-10.
16. AI-Detected Emotional State = Kondisi emosional pasien yang terdeteksi oleh AI (misalnya, Happy, Anxious).
17. Adherence to Treatment (%) = Persentase kepatuhan pasien terhadap rencana perawatan yang diberikan.

Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("mental_health_diagnosis_treatment_.csv")
df
```

```python
selected_features = [
    "Age", "Gender", "Diagnosis", "Symptom Severity (1-10)", "Mood Score (1-10)", 
    "Sleep Quality (1-10)", "Physical Activity (hrs/week)", "Medication", "Therapy Type",
    "Treatment Duration (weeks)", "Stress Level (1-10)", "Treatment Progress (1-10)",  
    "AI-Detected Emotional State", "Adherence to Treatment (%)"
]
df = df[selected_features]
```

```python
df.info()
```

Data Cleansing

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
numerical_features = ["Age", "Symptom Severity (1-10)", "Mood Score (1-10)", 
                      "Sleep Quality (1-10)", "Physical Activity (hrs/week)", 
                      "Treatment Duration (weeks)", "Stress Level (1-10)", 
                      "Treatment Progress (1-10)", "Adherence to Treatment (%)"]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_features])
plt.xticks(rotation=45)
plt.show()
```

```python
def remove_outliers(df, col_name):
    Q1 = df[col_name].quantile(0.25)
    Q3 = df[col_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df[(df[col_name] >= lower_bound) & (df[col_name] <= upper_bound)]
    return df_cleaned

df_cleaned = df.copy()

df_cleaned = remove_outliers(df_cleaned, 'Age')
df_cleaned = remove_outliers(df_cleaned, 'Adherence to Treatment (%)')

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_cleaned[["Age", "Adherence to Treatment (%)"]])
plt.xticks(rotation=45)
plt.show()
```

Pre-Procesing Data

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()

categorical_cols = ['Gender', 'Diagnosis', 'Medication', 'Therapy Type', 'AI-Detected Emotional State']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

Eksplorasi Data Analist

```python
numerical_features = ["Age", "Symptom Severity (1-10)", "Mood Score (1-10)", 
                      "Sleep Quality (1-10)", "Physical Activity (hrs/week)", 
                      "Treatment Duration (weeks)", "Stress Level (1-10)", 
                      "Treatment Progress (1-10)", "Adherence to Treatment (%)"]

korelasi_fitur = df[numerical_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(korelasi_fitur, annot=True, cmap="RdPu", fmt=".2f", linewidths=0.5)
plt.title("Korelasi Pearson Antar Fitur")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
df_baru = df.copy()

colom_pisah = ["Gender", "Diagnosis", "Medication", "Therapy Type", "AI-Detected Emotional State"]

standarisasi_kolom = [col for col in df_baru.columns if col not in colom_pisah]

scaler = StandardScaler()

df_baru[standarisasi_kolom] = pd.DataFrame(
    scaler.fit_transform(df_baru[standarisasi_kolom]),
    columns=standarisasi_kolom
)
```

```python
df_baru
```

Melatih Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
X = df_baru.drop(columns=["Adherence to Treatment (%)"])
y = df_baru["Adherence to Treatment (%)"]
```

```python
model_mlr = LinearRegression()
model_gbr = GradientBoostingRegressor(random_state=42)
```

```python
ration_test = [0.2, 0.3, 0.4]
```

Model Linear Regression

```python
for i in ration_test:
    print(f"\nEvaluasi Model dengan test_size = {i}")
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    model_mlr.fit(X_train, y_train)
    
    y_pred_train = model_mlr.predict(X_train)
    y_pred_test = model_mlr.predict(X_test)

    print(f"R2_score Training MLR: {r2_score(y_train, y_pred_train):.4f}")
    print(f"R2_score Testing MLR: {r2_score(y_test, y_pred_test):.4f}")
```

R2 Score yang sangat kecil (0.0150) = Model hampir tidak menjelaskan variasi dalam data training.

R2 Score negatif di data testing (-0.0049) = Model lebih buruk dibandingkan dengan asumsi rata-rata.

Model MSE (Mean Squared Error) & RMSE (Root Mean Squared Error)

```python
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)

print(f"MSE Training MLR       : {mse_train:.4f}")
print(f"MSE Testing MLR        : {mse_test:.4f}")
print(f"RMSE Training MLR      : {rmse_train:.4f}")
print(f"RMSE Testing MLR       : {rmse_test:.4f}")
print("-" * 50)
```

MSE dan RMSE di testing lebih tinggi dibanding training, menunjukkan bahwa model mungkin mengalami overfitting ringan.
    
RMSE lebih mudah diinterpretasikan dibanding MSE karena dalam satuan asli target.

Nilai RMSE mendekati 1, menunjukkan bahwa model masih memiliki error yang cukup besar.

Model GBR (Gradient Boosting Regressor)

```python
model_gbr.fit(X_train, y_train)

y_pred_train_gbr = model_gbr.predict(X_train)
y_pred_test_gbr = model_gbr.predict(X_test)

r2_train_gbr = r2_score(y_train, y_pred_train_gbr)
r2_test_gbr = r2_score(y_test, y_pred_test_gbr)

mse_train_gbr = mean_squared_error(y_train, y_pred_train_gbr)
mse_test_gbr = mean_squared_error(y_test, y_pred_test_gbr)
rmse_train_gbr = np.sqrt(mse_train_gbr)
rmse_test_gbr = np.sqrt(mse_test_gbr)

print(f"R2 Score Training GBR  : {r2_train_gbr:.4f}")
print(f"R2 Score Testing GBR   : {r2_test_gbr:.4f}")
print(f"MSE Training GBR       : {mse_train_gbr:.4f}")
print(f"MSE Testing GBR        : {mse_test_gbr:.4f}")
print(f"RMSE Training GBR      : {rmse_train_gbr:.4f}")
print(f"RMSE Testing GBR       : {rmse_test_gbr:.4f}")
print("-" * 50)
```

GBR overfitting pada dataset ini = performa training bagus, tapi prediksi pada data baru buruk.
